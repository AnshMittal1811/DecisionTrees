##### Treating Outliers in n_hot_rooms

```python
df.info()
```

```python
np.percentile(df.n_hot_rooms, [99])
```

```python
np.percentile(df.n_hot_rooms, [99])[0]
```

```python
ul = np.percentile(df.n_hot_rooms, [99])[0]
```

```python
df[(df.n_hot_rooms>ul)]
```

```python
df.n_hot_rooms[(df.n_hot_rooms> 3*ul)] = 3*ul
```

```python
df[(df.n_hot_rooms>ul)]
```

##### Treating Outliers in rainfall

```python
ll = np.percentile(df.rainfall, [1])[0]
```

```python
ll
```

```python
df[(df.rainfall < ll)]
```

```python
df.rainfall[(df.rainfall < 0.3*ll)] = 0.3*ll
```

```python
df[(df.rainfall < ll)]
```

##### Fill Missing values in n_hos_beds

```python
df.n_hos_beds = df.n_hos_beds.fillna(df.n_hos_beds.mean())
```

```python
df.info()
```

##### Variable Transformation in crime_rate

```python
sns.jointplot(x = "crime_rate", y = "price", data= df)
```

```python
df.crime_rate = np.log(1 + df.crime_rate**2)
```

```python
sns.jointplot(x = "crime_rate", y = "price", data= df)
```

##### Transforming distance variables

```python
df['avg_dist'] = (df.dist1 + df.dist2 + df.dist3 + df.dist4)/4
```
